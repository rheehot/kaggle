```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
train = pd.read_csv('./input/train.csv')
test = pd.read_csv('./input/test.csv')
```

```python
train.head()
```

```python
train = train.drop(['Cabin', 'Embarked', 'Name', 'Ticket'],axis=1)
test = test.drop(['Cabin', 'Embarked', 'Name', 'Ticket'],axis=1)
```

```python
train.head()
```

```python
test.head()
```

```python
train["Age"].fillna(train.groupby("Sex")["Age"].transform("mean"), inplace=True)
test["Age"].fillna(test.groupby("Sex")["Age"].transform("mean"), inplace=True)
test["Fare"].fillna(test.groupby("Age")["Fare"].transform("mean"), inplace=True)
```
